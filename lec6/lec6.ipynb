```python
import numpy as np
import torch
from torch.autograd import Variable as V
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from torch import FloatTensor as T
```

```python
ds=np.random.multivariate_normal([0,2],[[1.1,1],[1,1.1]],100)
x=ds[:,0]
y=ds[:,1]
plt.figure(figsize=(10,5))
plt.scatter(x,y)
plt.grid(1)
plt.show()
```

```python
def mse(y_pred,y_true):
    return ((y_pred-y_true)**2).mean()
```

```python
def line(x,m,b):
    return (x*m) + b
```

```python
def mse_loss(x,y,m,b):
    return mse(line(x,m,b),y)
```

```python
mse_loss(x,y,1,1)
```

```python
x=V(T(x))
y=V(T(y))
```

```python
m=V(torch.randn(1),requires_grad=True)
b=V(torch.randn(1),requires_grad=True)
```

```python
lr=1e-3
epochs=1000
```

```python
for e in range(epochs):
    loss=mse_loss(x,y,m,b)
    if (e%100):
        print (loss.data[0])
    
    loss.backward()
    m.data = m.data - (lr*m.grad.data)
    b.data = b.data - (lr*b.grad.data)
    
    m.grad.data.zero_()
    b.grad.data.zero_()
    
```

```python
a=line(range(-3,5),m.data.numpy()[0],b.data.numpy()[0])
print (a)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(ds[:,0],ds[:,1])
plt.plot(range(-3,5),a)
plt.grid(1)
plt.show()
```
